SYNCHING AUDIO TO FIT VIDEO

```python
txt_path = r"response_.txt"
output_audio_path = "Tutorial_audio_timed.mp3"

# Initialize the TTS engine
engine = pyttsx3.init()

# Change voice
voices = engine.getProperty("voices")
engine.setProperty("voice", voices[0].id)  # 0 for male, 1 for female

# Change speech rate
engine.setProperty("rate", 200)  # Set new rate (default is ~200)

# Adjust the speech volume
engine.setProperty("volume", 0.9)  # Set new volume (0.0 to 1.0)

# Define start times and durations for each line in seconds
start_times = [0, 1, 5, 10, 16, 21, 23, 28, 29, 31, 36, 42, 45, 48, 53, 55, 60, 63, 68, 74, 80]
durations = [1, 4, 5, 6, 5, 2, 5, 1, 2, 5, 6, 3, 3, 5, 2, 5, 3, 5, 6, 6]

try:
    with open(txt_path, "r", encoding="utf-8") as file:
        lines = [line.strip() for line in file if line.strip()]  # Remove empty lines

    # Ensure lines and durations match
    if len(lines) != len(durations):
        raise ValueError("Number of lines and durations do not match!")

    combined_audio = AudioSegment.silent(duration=0)  # Start with silence

    # Process each line with its corresponding duration
    for line, duration in zip(lines, durations):
        print(f"Speaking: {line} (Duration: {duration}s)")
        temp_wav_path = "temp_output.wav"

        # Save the line to a temporary file
        engine.save_to_file(line, temp_wav_path)
        engine.runAndWait()

        # Load the audio segment
        segment_audio = AudioSegment.from_file(temp_wav_path)

        # If the segment is shorter than the duration, add silence
        segment_duration = len(segment_audio) / 1000.0
        if segment_duration < duration:
            silence = AudioSegment.silent(duration=(duration - segment_duration) * 1000)
            segment_audio += silence

        # Append the segment to the combined audio
        combined_audio += segment_audio

        # Remove the temporary file
        os.remove(temp_wav_path)

    # Export the final combined audio
    combined_audio.export(output_audio_path, format="mp3")
    print(f"Audio saved to {output_audio_path}")

except FileNotFoundError:
    print(f"The file {txt_path} was not found.")
except Exception as e:
    print(f"An error occurred: {e}")

# Clean up resources
engine.stop()
ipd.Audio(output_audio_path)
```

SYNCHING AUDIO TO FIT VIDEO  OPTION II

```python
txt_path = r"response_.txt"

# Define start times and durations for each line in seconds
start_times = [0, 1, 5, 10, 16, 21, 23, 28, 29, 31, 36, 42, 45, 48, 53, 55, 60, 63, 68, 74, 80]
durations = [1, 4, 5, 6, 5, 2, 5, 1, 2, 5, 6, 3, 3, 5, 2, 5, 3, 5, 6, 6]


# Read the text file
with open(txt_path, "r", encoding="utf-8") as file:
    lines = [line.strip() for line in file if line.strip()]
     # Ensure lines and durations match
    if len(lines) != len(durations):
        raise ValueError("Number of lines and durations do not match!")

# Initialize pyttsx3 engine
engine = pyttsx3.init()

# Configure voice properties
engine.setProperty('voice', engine.getProperty('voices')[0].id)
engine.setProperty('rate', 200)

# Generate audio for each line
audio_segments = []
for idx, line in enumerate(lines):
    print(f"Processing line: '{line}' with duration {durations[idx]}s")
    
    # Generate spoken audio for the line
    temp_wav = f"line_{idx}.wav"
    try:
        # Save to temporary WAV file
        engine.save_to_file(line, temp_wav)
        engine.runAndWait()
        
        # Load the generated audio
        spoken_audio = AudioSegment.from_wav(temp_wav)
        
        # Calculate required silence to match the duration
        silence_duration = (durations[idx] * 1000) - len(spoken_audio)  # Convert to ms
        if silence_duration > 0:
            silence_segment = AudioSegment.silent(duration=silence_duration)
            full_segment = spoken_audio + silence_segment
        else:
            full_segment = spoken_audio[:durations[idx] * 1000]  # Trim excess audio

        # Append the processed segment
        audio_segments.append(full_segment)

    finally:
        # Remove the temporary WAV file
        if os.path.exists(temp_wav):
            os.remove(temp_wav)

# Concatenate all audio segments
total_duration = (start_times[-1] + durations[-1]) * 1000  # Total audio duration in ms
final_audio = AudioSegment.silent(duration=total_duration)  # Initialize with silence

for idx, segment in enumerate(audio_segments):
    start_time = start_times[idx] * 1000  # Convert to ms
    final_audio = final_audio.overlay(segment, position=start_time)

# Export the final audio
output_path = "final_tutorial_audio.mp3"
final_audio.export(output_path, format="mp3")

print(f"Audio saved to {output_path}")
engine.stop()
ipd.Audio("final_tutorial_audio.mp3")
```
