```python
# Investigate
vid_path = r"first_screen_record.mp4"

# Video specifications 
video_width, video_height, fps = 1280, 720, 24
codec = "libx264"

# Read the video
video = mpy.VideoFileClip(vid_path).resize(newsize=(video_width, video_height))

# Write the video
video.write_videofile("first_screen_record_compatible.mp4", codec=codec, fps=fps, preset="slow")

# Display the video
video.close()
ipd.Video("first_screen_record_compatible.mp4", width=650, height=350, embed=True)
```
